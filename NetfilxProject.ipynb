```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading CSV file
df=pd.read_csv("netflix.csv")
df
```

### Checking dataframe

```python
#Getting info about the Dataframe
df.info()
```

```python
df["director"].value_counts().head(10).plot(kind="bar")
# plt.xlabel("Content Type")
# plt.ylabel("Counts")
plt.title(" ")
```

```python

#Getting Statical Informaton about Dataframe
df.describe()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], infer_datetime_format=True, errors='coerce')
```

```python

constraint=df['cast'].apply(lambda x:str(x).split(',')).tolist()
constraint[2]
```

```python

df_new=pd.DataFrame(constraint,index=df['title'])
df_new.head()
```

```python
df_new=df_new.stack()
df_new.head(40)
```

```python
df_new=pd.DataFrame(df_new.reset_index())
df_new.rename(columns={0:'Actors'},inplace=True)
```

```python
df_final=df_new.merge(df[['show_id','type','director','country','listed_in','title','release_year','rating','duration']],on=['title'],how="left")
df_final["Actors"].value_counts().head().plot(kind="bar")
plt.xlabel("Actors")
```

```python
#Checking empty values in Dataframe
df.isna().sum()
```

```python
#Filling Empty Values in DataFrame
df=df.fillna({"director":"Not Known", "cast":"Not Known","country": "Not Known","rating":"unavilable"})
```

```python
df[df.date_added.isnull()]
```

There are 10 missing value in 'date_added' column so replacing it with mean date

```python
meandate=df["date_added"].mean()
df=df.fillna({'date_added':meandate})
df
```

```python
df[df.duration.isnull()]
```

```python
df[df.director =='Louis C.K.'].head()
```

Here we find that in three rows the rating is in mins (duration data mistyped into rating columns) so fixing it

```python
df.loc[df['director']=='Louis C.K.','duration']=df['rating']
df[df.director =='Louis C.K.'].head()
```

```python
df.loc[df['director']=='Louis C.K.','rating']="Unavilable"
```

### Analysing data and Visualizations


```python
#Analysing rating of content in Dataframe
df["rating"].value_counts().plot(kind="bar")
plt.xlabel("Rating")
plt.ylabel("count")
plt.title("Rating")
```

```python
#Most produced Genre 
df["listed_in"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Genre")
plt.title("Most produced Genre ")
```

```python
#Most Content Producing Countries 
df["country"].value_counts().head().plot(kind="bar")
plt.xlabel("Country")
plt.title("Most Content Producing Countries ")
```

```python
#Analysing which type of content in Dataframe
df["type"].value_counts().plot(kind="pie")
# plt.xlabel("Content Type")
# plt.ylabel("Counts")
plt.title("Content ")
```

```python
#most Content added yeare
df["date_added"].dt.year.value_counts().plot(kind="bar")
plt.xlabel("year")
```

```python
#Analysing which type of content in Dataframe in India
df_ind=df[df["country"]=="India"]
df_ind["type"].value_counts().plot(kind="pie")
```

```python
#Most Content added year wise in India
df_ind["date_added"].dt.year.value_counts().plot(kind="bar")
plt.xlabel("Year")
```

### Analysis of Indian Contant

```python
#Analysing rating of content in Dataframe in India
df_ind["rating"].value_counts().plot(kind="bar")
plt.xlabel("Rating")
plt.ylabel("count")
plt.title("Rating in india")
```

```python
#2018-present movie-TV show analysis
df2018=df[df["date_added"].dt.year<=2018]
df2018["type"].value_counts().plot(kind="pie")
```

```python
#Analysing movie release year wise
dfm=df[df["type"]=="Movie"]
dfm["release_year"].value_counts().head(20).plot(kind="bar")
plt.xlabel("Year")
plt.ylabel("Movie")
plt.title("Movie vs year")
```
